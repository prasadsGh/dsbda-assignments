```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Titanic-Dataset.csv")
```

```python
mean_age = df['Age'].mean()
df['Age'].fillna(value = mean_age, inplace = True)
```

```python
df2 = df.drop("Cabin", axis =1)
```

```python
sns.boxplot(data = df2, x= "Sex", y= "Age", hue = "Survived")
```

```python
sns.boxplot(data = df2, x= "Sex", y= "Survived")
```

```python
sns.boxplot(data = df2, x= "Survived", y= "Sex")
```
